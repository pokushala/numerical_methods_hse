## 1.Задать матрицу системы A и вектор правой части b.

Используя встроенную функцию, найти решение системы с помощью метода Гаусса

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

A = np.array([ [29.7,2,0,19.8,2], [9.9,-21,0,-9.9,1], [-9.9,11,29,6.6,1], [9.9,7.5,2,-19.8,0], [-49.5,-1,23,9.9,84] ])
b = np.array([26.2, -41.1, 97.4, 99.8, 27.1])
```

```python
x_solve = np.linalg.solve(A,b)
print('решение с помощью встроенной функции:\n x =', x_solve)
```

## 2. Преобразовать систему.
$Ax = b$ к виду $x = Bx+c$, удобному для итераций. Проверить выполнение достаточного условия сходимости итерационных методов $||B||_\infty <1$.

Систему удобно привести к виду:

$x_1 = \frac{1}{a_{11}}(b_1-a_{12}x_2-\dots-a_{1m}x_m)$

$\dots$

$x_m = \frac{1}{a_{mm}}(b_m-a_{m1}x_1-\dots-a_{m-1 m}x_{m-1})$

Получаем 


$b_{ij}= \begin{cases}
0, \ \ i=j\\
-\frac{a_{ij}}{a{ii}}, \ \ i \ne j
\end{cases}$

$c_i = \frac{b_i}{a_{ii}}$

$\overline{x}^{(n+1)}=B\overline{x}^{(n)}+\overline{c}$

```python
B = np.empty_like(A)
c = np.empty_like(b)

for i in range(len(c)):
    for j in range(len(c)):
        B[i,j] = 0 if i==j else -A[i,j]/A[i,i]
    c[i]=b[i]/A[i,i]
```

Проверим, что достаточное условие сходимости выполняется:

```python
B_norm = np.linalg.norm(B, np.inf)
print('||B|| =', B_norm,'< 1 => условие выполнено')
```

## 3. Bыполнить 10 итераций по методу Зейделя
взять любое начальное приближение. Принимая решение, полученное в п. 1 за точное, найти величину абсолютной погрешности итерационного решения (использовать норму $||.||_{\infty)}$.

```python
#использован алгоритм с семинара

def zeid(B, c, x, iterations):
    n = len(c)
    for iteration in range(iterations):
        x_new = np.copy(x)
        for i in range(n):
            x_new[i] = B[i,:i].dot(x_new[:i]) + B[i,i+1:].dot(x[i+1:]) + c[i]
        x = x_new
        print("Приближение", iteration + 1, "=", x)
    return x

print("Для начального приближения", np.zeros((len(c),)))
x1 = zeid(B, c, np.zeros((len(c),)), 10)
print("\nРешение методом Гаусса-Зейделя для 10 итераций:", x1)
print("Абсолютная погрешность решения: ", np.linalg.norm((x_solve - x1), np.inf))

print("\n\nДля начального приближения", np.ones((len(c),)))
x2 = zeid(B, c, np.ones((len(c),)), 10)
print("\nРешение методом Гаусса-Зейделя для 10 итераций:", x2)
print("Абсолютная погрешность решения: ", np.linalg.norm((x_solve - x2), np.inf))
```

### Метод Гаусса-Зейделя
Приближения $x^{k+1}$ вычисляются по формулам:

$B_1$ - нижнетреугольная матрица

$B_2$ - верхнетреугольная матрица

В силу $B = B_1 + B_2 \Rightarrow \bar{x}^{k+1}=B_1\bar{x}^{k+1}+B_2\bar{x}^{k}+\bar{c}$

Критерий окончания:

$||\bar{x}^{(n)}-\bar{x}^{(n-1)}||\le \frac{1-||B||}{||B_2||}\varepsilon$

## Задача 5.2. 
Для системы уравнений $Ax=b$ из задачи 5.1 найти решение по методу Зейделя с точностью $\varepsilon = 10^{-6}$, взяв любое начальное приближение. Для этого модифицировать функцию $zeid$ так, чтобы решение вычислялось с заданной точностью $\varepsilon$. Предусмотреть подсчет количества итераций, потребовавшихся для достижения точности.

```python
epsilon = 1e-6
```

```python
def zeid_norm(B, c, x, iterations, eps):
    n = len(c)
    B2 = np.copy(B)
    B2[np.tril_indices(len(B))] = 0
    k = (1 - np.linalg.norm(B, np.inf))/np.linalg.norm(B2, np.inf)
    
    for iteration in range(iterations):
        x_new = np.copy(x)
        for i in range(n):
            x_new[i] = B[i,:i].dot(x_new[:i]) + B[i,i+1:].dot(x[i+1:]) + c[i]
        if np.linalg.norm(x - x_new, np.inf) <= eps*k:
            return x_new, iteration + 1
        x = x_new
    return x, iteration + 1

x_norm, iter_norm = zeid_norm(B, c, np.zeros((len(c),)), 1000, epsilon)
print("Решение методом Зейделя с заданной точносью:", x_norm)
print("Число итераций:", iter_norm)
print("Абсолютная погрешность решения: ", np.linalg.norm((x_solve - x_norm), np.inf))
```

#B и c - матрица и вектор правой части соответствующей системы
#n - порядок матрицы B

#x0 - вектор начального приближения
#Возвращаемое значение - матрица k*n i-я строка это i-е приближение


def zeid(B, c, n, k, x0):
    res = np.zeros((k,n))
    y = x0
    for m in range(k):
        x = y
        for i in range(n):
            u = 0
            j = 0
            while 0<=j and j<i:
                #print(i)
                u = u + B[i,j]*y[j]
                j += 1
            j = i + 1
            while i<j and j<n:
                #print(i, j)
                u = u + B[i,j]*x[j]
                j += 1
            y[i] = u + c[i]
        for i in range(n):
            res[m,i] = y[i]
    return res
